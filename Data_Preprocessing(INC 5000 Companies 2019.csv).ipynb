## Handling Outliers :

```python
sns.boxplot(df)
```

```python
sns.boxplot(df['growth_%'])
```

```python
q1 = df['growth_%'].quantile(0.25)
q2 = df['growth_%'].quantile(0.50)
q3 = df['growth_%'].quantile(0.75)

IQR = q3 - q1
med = df['growth_%'].median()

upper_tail = q3 + 1.5 * IQR
lower_tail = q1 - 1.5* IQR


print(q1)
print(q2)
print(q3)
print(med)

print(upper_tail)
print(lower_tail)
```

```python
df['growth_%'].loc[df['growth_%']>upper_tail]
```

```python
df['growth_%'].loc[df['growth_%']>upper_tail]=upper_tail
```

```python
df['growth_%'].loc[df['growth_%']>upper_tail]
```
